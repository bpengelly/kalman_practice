```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline  
```

```python
def g_h_filter(data,x0,dx,g,h,dt):
    
    current_x=x0
    current_rate=h
    
    pred_vals=np.array([])
    
    for measurement in data:
        
        x_pred = current_x + dt*dx
        x_measured=measurement
        
        current_x= current_x+g*(x_measured-x_pred)
        
        current_rate= current_rate+h*(x_measured-x_pred)*1./dt
        
        pred_vals=np.append(pred_vals,current_x)
        
    return pred_vals
   
```

```python
weights = [158.0, 164.2, 160.3, 159.9, 162.1, 164.6, 
           169.6, 167.4, 166.4, 171.0, 171.2, 172.6]
rate=1.
time_step=1.

pred_weights=g_h_filter(weights,x0=160,dx=rate,g=6./10,h=2./3,dt=time_step)

print pred_weights
```

```python
def g_h_filter(data, x0, dx, g, h, dt=1., pred=None):    
    x = x0
    results = []
    for z in data:
        #prediction step
        x_est = x + (dx*dt)
        dx = dx        
        if pred is not None:
            pred.append(x_est)
        
        # update step
        residual = z - x_est
        dx = dx    + h * (residual) / dt
        x  = x_est + g * residual     
        results.append(x)  
    return np.array(results)

pred_weights=g_h_filter(weights,x0=160,dx=rate,g=6./10,h=2./3,dt=time_step)
```

```python

plt.figure(figsize=(13,6))
plt.plot(range(0,len(weights)),weights,label='Scale Measurements')
plt.plot(range(0,len(weights)),pred_weights,label='Filtered Measurements')
plt.scatter(range(0,len(weights)), [ 160+rate*i for i in range(0,len(weights))  ]  ,label='Actual Weight'     )
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
plt.show()
```

```python
from numpy.random import randn
def gen_data(x0, dx, count, noise_factor):
    return [x0 + dx*i + randn()*noise_factor for i in range(count)]
```

```python
dp=gen_data(5,2,100,noise_factor=100.)

rate=2.
preds=g_h_filter(dp,x0=5,dx=rate,g=2./10,h=0.01,dt=time_step)


plt.figure(figsize=(13,6))
plt.scatter(range(0,len(dp)),dp,label='Measurements')
plt.plot(range(0,len(dp)),preds,label='Filter')
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
plt.show()
```
